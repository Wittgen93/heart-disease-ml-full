# 01_preprocessing.ipynb

В этом ноутбуке:
- Загружаю train и test
- Очищаю их от невалидных значений и дубликатов
- Определяю и кодирую категориальные признаки
- Выровняю колонки test под train (без таргета)
- Сохраняю в data/processed/

```python
# 1. Чтение train и test + первичный обзор
df = pd.read_csv(f"{RAW_DIR}/train.csv")
df_test = pd.read_csv(f"{RAW_DIR}/test.csv")

print("Train shape:", df.shape)
print("Test shape :", df_test.shape)
print("Train columns:", df.columns.tolist())
print("Test columns :", df_test.columns.tolist())

print("\n--- TRAIN DTYPES ---")
print(df.dtypes)

print("\n--- TRAIN HEAD ---")
display(df.head())

print("\n--- TRAIN DESCRIBE ---")
display(df.describe())

print("\n--- TEST HEAD ---")
display(df_test.head())

print("\n--- TEST DESCRIBE ---")
display(df_test.describe())

# Кандидаты в категориальные признаки
thr = 10
cat_candidates = [c for c in df.columns if c not in ("ID","class") and df[c].nunique() <= thr]
print("\ncat candidates (<=10 uniq):", cat_candidates)
```

```python
# 2. Дубликаты/NA в train/test
df = df.drop_duplicates().dropna().reset_index(drop=True)
df_test = df_test.drop_duplicates().dropna().reset_index(drop=True)

print("after clean:", df.shape, df_test.shape)
```

```python
# 3. Приведение типов и диапазонов 
for d in (df, df_test):
    # дискретные числовые
    d["age"] = d["age"].round().clip(18, 120).astype("int16")
    d["resting_blood_pressure"] = d["resting_blood_pressure"].round().clip(60, 220).astype("int16")
    d["serum_cholestoral"] = d["serum_cholestoral"].round().clip(80, 600).astype("int16")
    d["maximum_heart_rate_achieved"] = d["maximum_heart_rate_achieved"].round().clip(50, 230).astype("int16")
    d["oldpeak"] = d["oldpeak"].clip(-1, 7).astype("float32")  # оставляем вещественным

    # категориальные
    d["chest"] = d["chest"].round().clip(1, 4).astype("int8")
    d["sex"] = d["sex"].clip(0, 1).astype("int8")
    d["fasting_blood_sugar"] = d["fasting_blood_sugar"].clip(0, 1).astype("int8")
    d["exercise_induced_angina"] = d["exercise_induced_angina"].clip(0, 1).astype("int8")
    d["resting_electrocardiographic_results"] = d["resting_electrocardiographic_results"].clip(0, 2).astype("int8")
    d["slope"] = d["slope"].clip(1, 3).astype("int8")
    d["number_of_major_vessels"] = d["number_of_major_vessels"].clip(0, 3).astype("int8")
    d["thal"] = d["thal"].where(d["thal"].isin([3, 6, 7]), 3).astype("int8")

print("types fixed:",
      df[["age","chest","resting_blood_pressure","serum_cholestoral",
          "maximum_heart_rate_achieved","oldpeak"]].dtypes.to_dict())
```

```python
# 4) Sanity-check после фиксов

print("DTYPES (key):", df[[
    "age","chest","resting_blood_pressure","serum_cholestoral",
    "maximum_heart_rate_achieved","oldpeak","sex","fasting_blood_sugar",
    "exercise_induced_angina","resting_electrocardiographic_results",
    "slope","number_of_major_vessels","thal","class"
]].dtypes.to_dict())

# Диапазоны числовых (train/test)
def rng(col, lo, hi):
    t_bad = (~df[col].between(lo, hi)).sum()
    s_bad = (~df_test[col].between(lo, hi)).sum()
    print(f"{col:>32}: train_min={df[col].min()} max={df[col].max()} | test_min={df_test[col].min()} max={df_test[col].max()} | out_of_range train={t_bad} test={s_bad}")

rng("age", 18, 120)
rng("resting_blood_pressure", 60, 220)
rng("serum_cholestoral", 80, 600)
rng("maximum_heart_rate_achieved", 50, 230)
rng("oldpeak", -1, 7)

# Категориальные
cats = {
    "sex": {0,1},
    "chest": {1,2,3,4},
    "fasting_blood_sugar": {0,1},
    "exercise_induced_angina": {0,1},
    "resting_electrocardiographic_results": {0,1,2},
    "slope": {1,2,3},
    "number_of_major_vessels": {0,1,2,3},
    "thal": {3,6,7},
}

for c, dom in cats.items():
    if c in df.columns:
        bad_tr = (~df[c].isin(dom)).sum()
        bad_te = (~df_test[c].isin(dom)).sum()
        print(f"{c:>32}: uniques_train={sorted(df[c].unique().tolist())} | bad train={bad_tr} test={bad_te}")
```

```python
# 5) Короткий обзор после фиксов
print("\n--- TRAIN HEAD (post-fix) ---")
display(df.head())

print("\n--- TRAIN DESCRIBE (post-fix) ---")
display(df.describe())

print("\n--- TEST HEAD (post-fix) ---")
display(df_test.head())

print("\n--- TEST DESCRIBE (post-fix) ---")
display(df_test.describe())
```

```python
# 6) OHE и выравнивание test под train (не включая target)
cat_cols = [
    "sex",
    "chest",
    "fasting_blood_sugar",
    "resting_electrocardiographic_results",
    "exercise_induced_angina",
    "slope",
    "number_of_major_vessels",
    "thal",
]

df = pd.get_dummies(df, columns=cat_cols, drop_first=True, dtype="int8")
df_test = pd.get_dummies(df_test, columns=[c for c in cat_cols if c in df_test.columns],
                         drop_first=True, dtype="int8")

feature_cols = [c for c in df.columns if c != "class"]
df_test = df_test.reindex(columns=feature_cols, fill_value=0)

print("final shapes:", df.shape, df_test.shape)
print("cols equal:", df_test.columns.tolist() == feature_cols)
```

```python
# 7) Сохранение в processed 
df.to_csv(f"{PROC_DIR}/train_preprocessed.csv", index=False)
df_test.to_csv(f"{PROC_DIR}/test_preprocessed.csv", index=False)
print("saved to:", PROC_DIR)
```

- привели типы и диапазоны 
- закодировали категориальные признаки через One-Hot Encoding (`drop_first=True`);
- выровняли столбцы `test` под схему `train` (без `class`).
